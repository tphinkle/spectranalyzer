```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

import struct

import alsaaudio
import time
import audioop
```

```python
inp = alsaaudio.PCM(alsaaudio.PCM_CAPTURE, alsaaudio.PCM_NONBLOCK)

inp.setchannels(1)
inp.setrate(8000)
inp.setformat(alsaaudio.PCM_FORMAT_S16_LE)

inp.setperiodsize(160)

```

```python
for i in range(10):
    l, data = inp.read()
    data_ = struct.unpack("160H", data)
    plt.plot(data_)
    plt.show()
    
    time.sleep(1)
```
